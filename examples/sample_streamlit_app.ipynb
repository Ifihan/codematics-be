# Sample Streamlit Application
This notebook demonstrates a simple Streamlit app that can be deployed to Cloud Run.

```python
import streamlit as st
import pandas as pd
import numpy as np
```

```python
st.title('Hello Cloud Run!')
st.write('This is a Streamlit app deployed from a Jupyter Notebook')
```

```python
data = pd.DataFrame({
    'x': np.arange(10),
    'y': np.random.randn(10)
})
```

```python
st.line_chart(data.set_index('x'))
```

```python
user_input = st.text_input('Enter your name:', 'World')
st.write(f'Hello, {user_input}!')
```
